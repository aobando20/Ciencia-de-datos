<h3 style="color:#76D7C4 "> 2.1 Contar los valores faltantes o perdidos por variable</h3>

```python
valoresperdidos=dfanime.isnull()
valoresperdidos.head()
```

```python
#Se verifica columna por columna los valores NAN
# Los valores TRUE son valores NAN

for column in valoresperdidos.columns.values.tolist():
    print(column)
    print(valoresperdidos[column].value_counts())
    print("")
```

```python
#Se reemplazan los valores Unknown por NAN
dfanime["Episodes"].replace("Unknown",np.nan,inplace=True)
```

<h3 style="color:#76D7C4 "> 2.2. Imputar valores numericos con la media</h3>

```python
dfanime
```

```python
avg_norm=dfanime["Rating"].mean(axis=0)
print("La media de perdidas normalizadas es:", avg_norm)
```

```python
dfanime["Rating"].replace(np.nan,6.47,inplace=True)
```

<h3 style="color:#76D7C4 "> 2.2. Imputar valores String con el mas frecuente</h3>

```python
dfanime["Genre"].value_counts().idxmax()
```

```python
dfanime["Genre"].replace(np.nan,"Hentai",inplace=True)
```

```python
dfanime["Type"].value_counts().idxmax()
```

```python
dfanime["Type"].replace(np.nan,"TV",inplace=True) 
```

```python
dfanime["Episodes"].value_counts().idxmax()
```

```python
dfanime["Episodes"].replace(np.nan,1,inplace=True)
```

```python
#Se verifica que no existan valores TRUE

valoresperdidos=dfanime.isnull()

for column in valoresperdidos.columns.values.tolist():
    print(column)
    print(valoresperdidos[column].value_counts())
    print("")
```

<h3 style="color:#45B39D"> Cuales son los generos de anime con rating mayor a 8? </h3>

```python
toprating=dfanime.loc[dfanime["Rating"]>8, ["Genre", "Rating"]]
toprating
```

```python
topgenre=toprating.groupby(dfanime["Genre"]).mean()
topgenre
```

```python
def filter_func(dfanime):
    return dfanime["Rating"].mean()>9.5

best=dfanime.groupby("Rating").filter(filter_func)
best
```

```python
best2=best.set_index("Anime_ID")
mapping={30120:"segundo mejor",33662:"el mejor"}
best2.groupby(mapping).sum()
```
