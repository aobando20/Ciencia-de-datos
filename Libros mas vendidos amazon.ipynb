<h3 style="color:#E7C910"> Cuales son los 10 Autores con mas reseñas en el 2019? </h3>

```python
año2019=dfamazon.loc[dfamazon.Year==2019,["Author","Name","Reviews"]]
año2019.head()
```

```python
año2019.describe()
```

```python
top_10 = año2019.sort_values('Reviews', ascending = False)[:10]
top_10
```

<h3 style="color:#E7C910"> Cuales son los 10 Autores con menos reseñas en el 2019? </h3>

```python
bottom_10 = año2019.sort_values('Reviews', ascending = True)[:10]
bottom_10
```

<h3 style="color:#E7C910"> Cuantas libros pertenecen a ciencia ficcion y cuales no? </h3>

```python
dfamazon.groupby(dfamazon["Year"])["Genre"].value_counts()
```

```python
genero=dfamazon.groupby(dfamazon["Year"])["Genre"].value_counts().reset_index(level=0).rename(columns={"Genre":"Genre","Genre":"Genre Count"})
genero
```

```python
genero.pivot_table("Genre Count",index="Genre",columns="Year")
```

```python
import matplotlib.pyplot as plt
import numpy as np
import math

labels = [2009,2010,2011,2012,2013,2014,2015,2016,2017,2018,2019] #Eje x

Fiction = [24,20,21,21,24,29,17,19,24,21,20] #Eje y abajo
Non_Fiction = [26,30,29,29,26,21,33,31,26,29,30] # Eje y arriba

width = 0.35       # Tamaño de barras

fig, ax = plt.subplots()

ax.bar(labels,Fiction , width, label='Fiction')
ax.bar(labels, Non_Fiction, width, bottom=Fiction,
       label='Non Fiction')

ax.set_ylabel('Generos')
ax.set_title('Cantidad de generos por año')
ax.legend()

plt.xticks (labels) # Sirve para colocar todos los valores del eje x
fig.set_figheight(5) # Sirve para cambiar el alto de la figura
fig.set_figwidth(8) # Sirve para cambiar el ancho de la figura

#Cambiar tamaño, ubicacion y sombra de leyenda
legend = plt.legend(loc='best', shadow=True, fontsize='10')
legend.get_frame().set_facecolor('white')
#best,upper right,upper left,lower left,lower right,right,center left,center right,lower center,upper center,center
plt.show()
```
