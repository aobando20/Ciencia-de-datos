```python
dfamazon['Date'] = pd.to_datetime(dfamazon.Date, format='%Y-%m-%d')
```

```python
import datetime

dfamazon["Year"]=dfamazon["Date"].dt.year
dfamazon["Month"]=dfamazon["Date"].dt.month
dfamazon["Day"]=dfamazon["Date"].dt.day
dfamazon
```

```python
años=dfamazon.iloc[:,[7,6]]
años
```

```python
grupoaños=años.groupby(dfamazon["Year"]).sum()
grupoaños             
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for plotting

x = [1997,1998,1999,2000,2001,2002,2003,2004,2005,2006,2007,2008,2009,2010,2011,2012,2013,
        2014,2015,2016,2017,2018,2019,2020,2021]

y = grupoaños["Volume"].tolist()

fig, ax = plt.subplots()
ax.plot(x, y)

plt.xticks (x) # Sirve para colocar todos los valores del eje x
fig.set_figheight(6) # Sirve para cambiar el alto de la figura
fig.set_figwidth(15) # Sirve para cambiar el ancho de la figura

ax.set(xlabel='Años', ylabel='Volumen de acciones',
       title='Volumen total de acciones por año')
ax.grid()


fig.savefig("test.png")
plt.show()
```
